The aim of this project is to clean the data and analyze the used car listings.

```python
import pandas as pd
import numpy as np
autos = pd.read_csv("autos.csv" , encoding = "Latin-1")
```

```python
autos.info()
autos.describe()
autos.head()
```

From the info obtained above, we can make the conclusion that the dataset has 50,000 rows
and 20 columns . Also we are able to take a quick peak at the first 5 rows of the data and 
now we understand how the data is organized. We also notice that most of the columns are strings and some of the columns have null values(not more then 20% of the values in the columns are null)

```python
autos.columns
```

```python
columns = autos.columns
modified_cols = []
for col in columns:
    col = col.replace("yearOfRegistration","registration_year")
    col = col.replace("monthOfRegistration","registration_month")
    col = col.replace("notRepairedDamage","unrepaired_damage")
    col = col.replace("dateCreated","ad_created")
    col = col.lower()
    modified_cols.append(col)
print(modified_cols)
autos.columns = modified_cols
print(autos.columns)
autos.head()
```

We have succesfully changes the column names from camel to snake case for visibility and replaced some of the column names that were vagues with more descriptive titles.

Now we want to look for columns where most of the values are the same. These columns will not provide much value for data exploration. A few candidates for removal include the seller and offertype columns.

```python
autos.describe()
autos.head()
```

Another note is that price and odometer columns are integers stored as strings.
We want to remove all non-numeric characters, convert the 2 columns to integers and also rename the odometer column name.

```python
vals = autos['price']
vals = vals.str.replace("$","").str.replace(",","").astype(int)
autos['price'] = vals
odo_vals = autos['odometer']
odo_vals = odo_vals.str.replace("km","").str.replace(",","").astype(int)
autos['odometer'] = odo_vals
autos.rename(columns = {'odometer':'odometer_km'} , inplace = True)
autos.head()
```

We'll continue exploring the data as we look for columns that inaccurate data. Let's start with the price and odometer_km columns

```python
prices = autos['price']
print(prices.unique().shape)
print(prices.describe())
prices.value_counts().head()
prices = prices.sort_values(ascending = False)
print(prices.value_counts())
```

```python
odo = autos['odometer_km']
print(odo.unique().shape)
print(odo.describe())
odo.value_counts().head()
odo = odo.sort_values(ascending = False)
print(odo.value_counts())
```

From the registration year stats, we notice a few odd values such as  the maximum year the car was registered as 999 and the minimum as 1000. Obviously, this is inaccurate based on the ad_created and date crawled dates entered. Therefore we'll analyze the column and decide what years are acceptable as the maximum and minimum registration years.

```python
sorted_vals = autos['registration_year'].sort_values(ascending = False)
reg_years  = dict(autos['registration_year'].value_counts())
```

We want to determine the percentage of our rows that have invalid registration data in it.Then, we'll decide whether to remove these rows depending on the percentage.

```python
invalid_sum = (~autos['registration_year'].between(1900,2016)).sum()
invalid_percent = invalid_sum/autos.shape[0]
print(invalid_percent)
```

Removing the invalid rows:

```python
autos = autos[autos['registration_year'].between(1900,2016)]
autos['registration_year'].value_counts(normalize=True)
```

Next, our goal is to explore the car prices by brand and decide which one is the most affordable vs expensive.
We'll achieve this by exploring the brand column.

```python
autos['brand'].value_counts(normalize = True)
```

German manufacturers represent four out of the top five brands, almost 50% of the overall listings. Volkswagen is by far the most popular brand, with approximately double the cars for sale of the next two brands combined.

There are lots of brands that don't have a significant percentage of listings, so we will limit our analysis to brands representing more than 5% of total listings.

```python
brand_freq = autos['brand'].value_counts(normalize = True)
common_brands = brand_freq[brand_freq > 0.05].index
print(common_brands)
```

```python
brand_avg_prices = {}
for brand in common_brands:
    brand_only = autos[autos['brand'] == brand]
    mean_price = brand_only['price'].mean()
    brand_avg_prices[brand] = int(mean_price)
print(brand_avg_prices)
```

```python
brand_mean_mileage = {}
for brand in common_brands:
    brand_only = autos[autos['brand'] == brand]
    mean_mileage = brand_only['odometer_km'].mean()
    brand_mean_mileage[brand] = int(mean_mileage)
print(brand_mean_mileage)
```

```python
bap_series = pd.Series(brand_avg_prices)
bmm_series = pd.Series(brand_mean_mileage)

df = pd.DataFrame(bap_series, columns=['mean_price'])
df['mean_mileage'] = bmm_series
df
```

The range of car mileages does not vary as much as the prices do by brand, instead all falling within 10% for the top brands. There is a slight trend to the more expensive vehicles having higher mileage,
with the less expensive vehicles having lower mileage.
